Feature Extraction/Featurneering -> pe Engiandas

Feature engineering means creating new columns or transforming existing ones to help the model perform better.

```python
def get_runsremark(data):
  if data > 500:
    return "Execellent Run"
  elif data > 300:
    return "Moderate Run"
  elif data > 0:
    return "Poor Run"
  else:
    return "Nothing"
```

```python
dataset['RunsRemark'] = dataset['Runs'].apply(get_runsremark)  # Creates a new column 'RunsRemark' by applying the function 'get_runsremark' to each value in the 'Runs' column
```

```python
dataset.head()
```

```python
def add_5_runs(data):
  data += 5
  return data
```

```python
dataset['Runs'] = dataset['Runs'].apply(add_5_runs)# Adds 5 to each value in the 'Runs' column using the 'add_5_runs' function
```

iloc stands for "integer location". It is used to access rows and columns by position/index number, not by label.

df.iloc[row_index, column_index]

```python
#. Access a specific row (by number)
dataset.iloc[0]    # First row
```

```python
dataset.iloc[2, 1]   # Row 2, Column 1 (3rd row, 2nd column)
```

```python

dataset.iloc[0, [1, 2]]   # First row, 2nd and 3rd columns
```

| Feature  | `loc`                              | `iloc`                              |
| -------- | ---------------------------------- | ----------------------------------- |
| Uses     | Labels (column names, index names) | Integer index positions (0,1,2,...) |
| Example  | `df.loc[2, 'Runs']`                | `df.iloc[2, 1]`                     |
| Safe for | Named access, filtering            | Positional access                   |

.loc[] is used to access rows and columns in a DataFrame by label (not position).
General Syntax:

df.loc[row_label, column_label]
row_label: index (or list of indices) you want to select.

column_label: column name (or list of names) to return.
--------------------------------------------------------------
# Get the index of the row where each team's player has the maximum Runs
max_run_indices = dataset.groupby('Team')['Runs'].idxmax()

# Use those indices to get the corresponding player names, team, and runs
top_scorers = dataset.loc[max_run_indices, ['Player Name', 'Team', 'Runs']]

# Sort by Runs in descending order
top_scorers = top_scorers.sort_values(by='Runs', ascending=False)


ðŸ§  Explanation:
groupby('Team')['Runs'].idxmax()
â†’ Finds the index of the player with the highest runs in each team.

dataset.loc[...]
â†’ Extracts full row details (player name, team, runs).

sort_values(by='Runs', ascending=False)
â†’ Ranks all top scorers from highest to lowest.

Returns the top run-scorer from each team, as a filtered table.

```python
dataset.loc[dataset.groupby('Team')['Runs'].idxmax(), ['Player Name', 'Team', 'Runs']].sort_values(by='Runs', ascending=False)

# What this does:
# groupby('Team')['Runs'].idxmax() â†’ Gets the index of the row with max runs per team

# loc[...] â†’ Selects those rows from the DataFrame

# [['Player Name', 'Team', 'Runs']] â†’ Picks relevant columns

# .sort_values(by='Runs', ascending=False) â†’ Sorts by highest runs

#-----------------------------------------------------------------------

# âœ… What .idxmax() means:
# .idxmax() returns the index (row label) of the maximum value in a Series or within each group in a groupby().

# dataset.groupby('Team')['Runs'].idxmax()
# This returns, for each team, the row index of the player who has the highest runs.

#----------------------------------------------------------------------------
# dataset.groupby('Team')['Runs'].idxmax()
# This returns the index of the row where each team's player scored the maximum runs.

# dataset.loc[...]
# # Now we use .loc[...] to fetch the full data from those rows.
# Means:

# Fetch only the rows with max-run players

# And from those rows, return only the 'Player Name', 'Team', and 'Runs' columns


```
