```python
import matplotlib.pyplot as plt
import numpy as np

# Create x values from 0 to 10
x = np.arange(0, 11, 1)   # [0, 1, 2, ..., 10]
y = x ** 2                # y = x²

# Plot
plt.plot(x, y, color='blue', linestyle='-', label='y = x²')

# Add title and labels
plt.title('Quadratic Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add grid
plt.grid(True)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create x values from 0 to 2π
x = np.linspace(0, 2 * np.pi, 100)

# Calculate y values
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot sin(x)
plt.plot(x, y_sin, color='blue', linestyle='-', label='sin(x)')

# Plot cos(x)
plt.plot(x, y_cos, color='red', linestyle='--', label='cos(x)')

# Add title and labels
plt.title('Sine and Cosine Functions')
plt.xlabel('x')
plt.ylabel('y')

# Add legend
plt.legend()

# Add grid
plt.grid(True)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create x values from 0 to 2π
x = np.linspace(0, 2 * np.pi, 100)

# Calculate y values
y_sin = np.sin(x)
y_cos = np.cos(x)

# Create figure
plt.figure(figsize=(12, 5))

# 1st subplot: sin(x)
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st plot
plt.plot(x, y_sin, color='blue')
plt.title('sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)

# 2nd subplot: cos(x)
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd plot
plt.plot(x, y_cos, color='red')
plt.title('cos(x)')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.grid(True)

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
countries = ['USA', 'India', 'China', 'Brazil', 'Germany']
population = [331, 1391, 1444, 213, 83]  # Example population in millions

# Create figure
plt.figure(figsize=(8,6))

# Define hatch patterns for each bar
hatches = ['/', '\\', 'x', '+', '.']

# Plot bars with custom hatches
bars = plt.bar(countries, population, color='skyblue')

# Apply hatch to each bar
for bar, hatch in zip(bars, hatches):
    bar.set_hatch(hatch)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             height + 10,
             f'{height}M',
             ha='center',
             va='bottom')

# Add title and Y-axis label
plt.title('Population of 5 Countries')
plt.ylabel('Population (in millions)')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate 1000 random numbers (standard normal distribution)
data = np.random.randn(1000)

# Create histogram with 20 bins
plt.figure(figsize=(8,6))
plt.hist(data, bins=20, color='skyblue', edgecolor='black')

# Add title and axis labels
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Add grid
plt.grid(axis='y')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate 100 random x and y values between 0 and 1
x = np.random.rand(100)
y = np.random.rand(100)

# Compute distance from origin for each point
distance = np.sqrt(x**2 + y**2)

# Create scatter plot with color mapping
plt.figure(figsize=(8,6))
scatter = plt.scatter(x, y, c=distance, cmap='viridis')

# Add colorbar
plt.colorbar(scatter, label='Distance from Origin')

# Add title and axis labels
plt.title('Scatter Plot Colored by Distance from Origin')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```
