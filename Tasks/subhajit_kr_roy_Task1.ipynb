```python
import numpy as np

#  Create Arrays

# 1D array of 20 random integers from 1 to 100
arr1d = np.random.randint(1, 101, 20)
print("1D Array:", arr1d)

# 5×5 matrix with values from 1 to 25
matrix = np.arange(1, 26).reshape(5, 5)
print("\n5x5 Matrix:\n", matrix)

# Reshape and Slicing

# Reshape 1D array to 4×5 matrix
matrix_4x5 = arr1d.reshape(4, 5)
print("\nReshaped 4x5 Matrix:\n", matrix_4x5)

# Extract 2nd and 4th columns
cols_2_4 = matrix_4x5[:, [1, 3]]
print("\n2nd and 4th Columns:\n", cols_2_4)

#  Mathematical Operations

mean_val = matrix_4x5.mean()
median_val = np.median(matrix_4x5)
max_val = matrix_4x5.max()
std_val = matrix_4x5.std()

print(f"\nMean: {mean_val:.2f}, Median: {median_val}, Max: {max_val}, Std Dev: {std_val:.2f}")

# Replace all values divisible by 5 with -1
matrix_4x5[matrix_4x5 % 5 == 0] = -1
print("\nMatrix after replacing values divisible by 5:\n", matrix_4x5)

#  Boolean Masking

# Get all values greater than mean
greater_than_mean = matrix_4x5[matrix_4x5 > mean_val]
print("\nValues greater than mean:", greater_than_mean)

# Count values less than 10
count_less_than_10 = np.sum(matrix_4x5 < 10)
print("\nCount of values less than 10:", count_less_than_10)
```

```python
#  IMPORT LIBRARIES
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Optional for heatmap

#  Load the cleaned DataFrame (assume it's already cleaned from Section 2)
df = pd.read_csv('iris.csv')

# -------------------------------------------
# LINE PLOT: Average sepal length per species
avg_sepal = df.groupby('species')['sepal_length'].mean()

plt.figure(figsize=(8,5))
plt.plot(avg_sepal.index, avg_sepal.values, marker='o', color='blue', label='Avg Sepal Length')
plt.title('Average Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average Sepal Length')
plt.grid(True)
plt.legend()
plt.show()

# -------------------------------------------
#  BAR PLOT: Count of each species
species_counts = df['species'].value_counts()

species_counts.plot(kind='bar', color=['red', 'green', 'blue'])
plt.title('Count of Each Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.grid(axis='y')
plt.show()

# -------------------------------------------
# SCATTER PLOT: Sepal Length vs Petal Length, colored by species
plt.figure(figsize=(8,6))

for species in df['species'].unique():
    subset = df[df['species'] == species]
    plt.scatter(subset['sepal_length'], subset['petal_length'], label=species)

plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.legend()
plt.grid(True)
plt.show()

# -------------------------------------------
#  HISTOGRAM: Sepal Width
plt.figure(figsize=(8,5))
plt.hist(df['sepal_width'], bins=10, color='purple', edgecolor='black')
plt.title('Histogram of Sepal Width')
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()

# -------------------------------------------
# BOXPLOT: Compare Sepal Length across species
plt.figure(figsize=(8,5))
sns.boxplot(x='species', y='sepal_length', data=df, palette='Set2')
plt.title('Boxplot of Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.grid(True)
plt.show()

# -------------------------------------------
#  HEATMAP: Correlation Matrix
corr = df.corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# -------------------------------------------
#  EXPORT: Save cleaned DataFrame
df.to_csv('iris_cleaned.csv', index=False)
print("Cleaned DataFrame saved as 'iris_cleaned.csv'.")
```
